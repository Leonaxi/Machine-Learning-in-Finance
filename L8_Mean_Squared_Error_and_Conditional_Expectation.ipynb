```
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-2, 2, 10000)

def f(x):
  return 1 + x**2

eps = np.random.normal(size=x.size)
```

```
y = f(x) + eps
```

```
# plt.scatter(x, f(x), s=1)
plt.scatter(x, y, s=1)
```

```
ones = np.ones_like(x)
X = np.column_stack([ones, x, x**2])
Theta = np.linalg.lstsq(X, y)[0]
```

```
predicted = X @ Theta

plt.figure(figsize=[10, 5])
plt.scatter(x, f(x), s=3)
# plt.scatter(x, y, s=1)
plt.scatter(x, predicted, s=.1, color='r')
```
